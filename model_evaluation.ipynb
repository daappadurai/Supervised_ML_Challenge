```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.ExcelFile("model evaluation.xlsx")
df.sheet_names
```

```python
df2 = pd.read_excel(df, sheet_name="Sheet1")
df2.head(6)
```

```python
df2.columns
```

```python
df2.plot(x="Model", y="Accuracy", kind="bar", color = "magenta")
plt.xlabel("Model")
plt.ylabel("Overall Model Accuracy Score")
plt.savefig("Accuracy.png")
plt.show()

```

```python
df2.plot(x="Model", y="precision (0)", kind="bar", color = "darkgreen")
plt.xlabel("Model")
plt.ylabel("Fraud detection Precision score")
plt.savefig("Precision.png")
plt.show()
```

```python
df2.plot(x="Model", y="Sensitivty (0)", kind="bar", color = "purple")
plt.xlabel("Model")
plt.ylabel("Fraud detection Sensitivity score")
plt.savefig("Sensitivity.png")
plt.show()
```

```python
df2.columns
```

```python
df2.plot(x="Model", y="f1 ratio (0)", kind="bar", color = "red")
plt.xlabel("Model")
plt.ylabel("Hormonic Mean f1-score")
plt.savefig("F1_ratio.png")
plt.show()
```

```python
df2.plot(x="Model", y="Fn", kind="bar", color = "purple", label="Number of Cases Wrongly Flagged as Risky")
plt.xlabel("Model")
plt.ylabel("Number of Cases")
plt.legend
plt.savefig("False_negative.png")
plt.show()
```
